```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(123471)
```

```python
atletas = pd.read_csv("athletes.csv")
atletas.head()
```

### A final do basquete masculino das Olimpíadas de 2016 foi entre Estados Unidos e Sérvia, em que os Estados Unidos ganharam de 96 a 66. Sabe-se que a altura é um parâmetro bem relevante nesse esporte. Pode-se afirmar que os jogadores dessas duas seleções possuem mesma média de altura?

```python
final_basquete = atletas[atletas["nationality"].isin(["USA", "SRB"])] 
final_basquete = final_basquete[final_basquete["sex"] == "male"]
final_basquete = final_basquete[final_basquete["sport"] == "basketball"]
final_basquete
```

## Primeiro, precisa-se encontrar a estatística de teste

```python
def calcularEstatistica(data):
    data_auxiliar = data.groupby('nationality').mean()
    resultado = data_auxiliar["height"][0] - data_auxiliar["height"][1]
  
    return resultado
```

```python
est = round(calcularEstatistica(final_basquete), 5)
print(est)
```

## Nossa hipótese nula será que não há diferença de altura significativa entre os times


$$H_0: \mu_{T} = \mu_{\bar{T}}$$

$$H_1: \mu_{T} \neq \mu_{\bar{T}}$$

## Para calcular se há diferença significativa entre a altura dos jogadores dos dois times, pode-se utilizar um Teste de Permutação. 
## Consideraremos um nível de significância de 2%.

```python
def calcularTeste(data, significancia):
    filtro = data["nationality"] == "USA"
    diferencas = np.zeros(10000)
    
    for i in range(10000):    
        np.random.shuffle(filtro.values)
        diferencas[i] = data[~filtro]["height"].mean() - data[filtro]["height"].mean()
        
    LI = np.percentile(diferencas, (significancia/2)*100)
    LS = np.percentile(diferencas, (1 -significancia/2)*100)
        
    return diferencas, LI, LS
```

```python
vect, lim_inf, lim_sup = calcularTeste(final_basquete, 0.02)
print(lim_inf, " ", lim_sup)
```

## Gráficos podem ser gerados para melhor visualização

```python
plt.hist(vect, bins=20, rwidth=0.95)
plt.fill_between([lim_inf, lim_sup], 4, 2000, color='grey', alpha=0.8)
plt.axvline(x=est, color = "red")
plt.show
```

## Com os limites inferior, superior e a estatística de teste, pode-se responder a pergunta inicial

```python
def resultado(li, ls, est):
    if li < est and ls > est:
        return True
    return False
```

```python
result = resultado(lim_inf, lim_sup, est)
print(result)
```
